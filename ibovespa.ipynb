```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

import source.helpers as hp
```

```python
plt.style.use('ggplot')
```

```python
ibvsp_2019 = yf.download('^BVSP', start='2019-01-01', end='2019-12-31', progress=False)
ibvsp_2020 = yf.download('^BVSP', start='2020-01-01', end='2020-12-31', progress=False)
```

```python
ibvsp_2019.columns = [hp.snake_case(column) for column in ibvsp_2019.columns]
ibvsp_2020.columns = [hp.snake_case(column) for column in ibvsp_2020.columns]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,5), sharey=True)

years = (
    (ibvsp_2019, '2019', 'blue'),
    (ibvsp_2020, '2020', 'red'),
)

for i, (df, title, c) in enumerate(years):
    df.adj_close.plot(ax=ax[i], color=c)
    ax[i].title.set_text(title)
    ax[i].set_ylabel('Pontos')
    ax[i].set_xlabel('Data')
    
fig.suptitle("ndice Bovespa (Grfico 1)", fontsize=14)
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
